```python
# 1. Прочтите этот набор данных по пауэрлифтингу в pandas DataFrame.

import pandas as pd
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-08/ipf_lifts.csv"
df = pd.read_csv(url)
```

```python
# 2. Найдите рекорды каждого пола и каждого подразделения в каждом из 3 упражнений (жим лежа, приседания и становая тяга)
max_records = df.groupby(['sex', 'division'])[['best3bench_kg', 'best3squat_kg', 'best3deadlift_kg']].max()

print(max_records)
```

```python
"""3. Посчитайте количество побед каждого участника, принимая за победу 1-е
место. Сохраните результаты во вторичном DataFrame."""

# Створюємо новий DataFrame для збереження результатів
results_df = pd.DataFrame(columns=['name', 'wins', 'sex', 'division'])

# Перетворюємо тип даних стовбця "place" в числовий формат
df['place'] = pd.to_numeric(df['place'], errors='coerce')

# Здійснюємо групування даних за іменем учасника і підраховуємо кількість перемог
wins_count = df.loc[df['place'] == 1, ['name', 'sex', 'division']].value_counts().reset_index()
wins_count.columns = ['name', 'sex', 'division', 'wins']

# Заносимо результати в новий DataFrame
results_df = pd.concat([results_df, wins_count], ignore_index=True)

print(results_df)
```

```python
"""4. Используйте DataFrame из упражнения 3, чтобы показать для каждой
комбинации подразделения и пола, участника с наибольшим количеством побед
среди участников, когда-либо участвовавших в этом дивизионе"""

# Створюємо DataFrame для збереження результатів
winners_df = pd.DataFrame(columns=['division', 'sex', 'name', 'wins'])

# Групуємо дані по комбінації "division" и "sex"
grouped_df = results_df.groupby(['division', 'sex'])

# Визначаємо максимальну кількість перемог для кожної групи
max_wins = grouped_df['wins'].max()

# Визначаємо рядки з максимальним значенням перемог для кожної групи
for (division, sex), max_win in max_wins.items():
    winners = results_df[(results_df['division'] == division) & (results_df['sex'] == sex) & (results_df['wins'] == max_win)]
    winners_df = pd.concat([winners_df, winners], ignore_index=True)

print(winners_df)
```
